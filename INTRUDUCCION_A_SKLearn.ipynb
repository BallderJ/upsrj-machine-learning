#Librerias
Se preparan las herramientas necesarias

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

#Definicion de los datos
Se definen los valores de entrada, se hace un ajuste especia(Reshape) para convertir una lista simple en una columna y la varialbe dependiente son los resultados reales que corresponden a cada x.


```python
#y= m* x+b

#Variable independiente
# SciKit-learn requiere que se usen arreglos de 20
X= np.array([1,2,3,4,5]).reshape(-1,1)
print(X)

#Variable dependiente

y= np.array([2,4,5,4,5])
print(y)


```

#Visualizacion de los puntos
El codigo genera una grafica con puntos azules. Esto sirve para ver como estan distribuidos los datos y confirmar si parecen seguir una tendencia de lina recta.

```python
plt.scatter(X,y)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Dtos originales")
plt.grid(True)
plt.show()
```

#Creacion y entrenamiento del modelo
El codigo crea un objeto llamado model y luego usa una funcion llamada fit(ajustar). En este paso, la computadora analiza los datos y busca matematicamente la linea recta.

```python
#Generamos nuestro modelo de regresion linear
model= LinearRegression()
#alimentamos nuestro modelo

model.fit(X,y)

#parametros:m(peso), b(sesgo)

m=model.coef_[0]
b = model.intercept_

print(f"Pendiente (m):{m:2f}")
print(f"Intercepto(b):{b:2f}")
```

#Evauacion del modelo
Una vez que el modelo aprendio, se le pide que intente adivinar los resultados. **Mean Squared Error** nos dice que tan lejos quedaron.
**R2 Score** Es un puntaje de presicion del 0 al 1

```python
#Generamos una prediccion
y_prend =model.predict(X)

#Evaluamos esa prediccion
mse=mean_squared_error(y,y_prend)
r2=r2_score(y,y_prend)
print(f"Mean Squared Error:{mse:2f}")
print(f"R2 Score:{r2:2f}")
```

#Grafica de resultados finales
El codigo dibuja todo junto, pone los puntos azules originales y encima traza una **linea roja** Esta linea roja es la prediccion o el camino que el modelo considera más acertado para describir la relación entre los datos.

```python
plt.scatter(X,y,label="Datos reales")
plt.plot(X,y_prend,color="red",label="Modelo lineal")

#
plt.xlabel("X")
plt.ylabel("y")
plt.title("Regresion Lineal simple")
plt.grid(True)
plt.show()
```
