#Medición de tiempo de ejecución y rendimiento
Este código mide el rendimiento del procesador al calcular el tiempo exacto que tarda en ejecutar un bucle de gran escala: primero genera una lista con 100 millones de números, activa un cronómetro de alta precisión, recorre cada elemento realizando dos operaciones de suma por cada vuelta y, finalmente, detiene el cronómetro para mostrar en pantalla cuántos segundos necesito.

```python
lista = list(range(100000000))
inicio = time.perf_counter()
suma=0
for i in lista:
  suma += i
  suma += 1

final = time.perf_counter()

tiempo = final - inicio
print(f"Tiempo de ejecucion: {tiempo:.6f}sec")
```

#Medición del rendimiento de procesamiento

Este código calcula el tiempo exacto que le toma a la computadora procesar una lista de 100 millones de números, iniciando un cronómetro antes de realizar una suma masiva mediante un bucle y deteniéndolo al finalizar para mostrar en pantalla cuántos segundos tardó el procesador en completar todas las operaciones.

```python
np_lista = np.arange (100_000_000)
inicio = time.perf_counter()
suma_np=np.sum(np_lista)
final = time.perf_counter()
tiempo = final - inicio
print(f"Tiempo de ejecucion: {tiempo:.6f}sec")
```

#Consulta estructurada de datos mediante mapeo de índices

Este código guarda información como si fuera una tabla, cada fila es una persona y cada columna es un tipo de dato, como el nombre o el animal favorito.
En lugar de buscar los datos por números (fila 1, columna 2), el código usa nombres para encontrarlos.
Así, cuando se pide el animal favorito de Kevin, el programa sabe exactamente dónde buscar y muestra la respuesta sin que tengamos que contar filas o columnas.

```python
base_de_datos = [
["Oscar", 21, "Gatos"],
["Paula", 21, "Tiburones"],
["Jonatan", 24, "Hurones"],
["Kevin", 22, "Periquitos"],
["Lucero", 21, "Perros"],
["Israel", 21, "Lobos"],
["Sergio", 21, "Perros"],
["Vahir", 21, "Leones"],
["Diego", 21, "Pinguinos"]
]

columnas = {"Nombre":0, "Edad":1, "Animal":2}
filas    = {"Oscar":0, "Paula":1, "Jonatan":2, "Kevin":3, "Lucero":4, "Israel":5, "Sergio":6, "Vahir":7, "Diego":8}
print(base_de_datos[filas["Kevin"]][columnas["Animal"]])
```

#Manejo y filtrado de datos con la librería Pandas
Este código toma la información y la acomoda en una tabla ordenada, donde cada dato tiene su propio nombre (como “Nombre” o “Animal”).
Se puede buscar directamente a Kevin por su nombre, sin contar filas ni posiciones, y el programa muestra automáticamente qué animal es el suyo.

```python
import pandas as pd
df = pd.DataFrame(base_de_datos, columns=["Nombre", "Edad", "Animal"], index = range(1,10))

print(df.Animal[df['Nombre']== "Kevin"])
```
