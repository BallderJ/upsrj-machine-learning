#Implemetacion de Pipeline

Un Pipeline encapsula todo el flujo de procesamiento de datos en un solo objeto.

Para el usuario, entrenar un pipeline es igual que entrenar un modelo

El escalonado no cambia el resultado conceptual, pero prepara el terreno para modelos mas complejos.


#JUSTIFICACION

Hasta ahora hemos entrenado modelos pasao por paso.
En la vida real esos paso siempre se repiten  en el mismo orden.
Un pipeline es una forma de decirle a la computadora: **"Estos son los pasos, siempre ejecutalos asi.**

```python
# 1. Definiremos las variables dependientes e independientes de nuestro modelo.
# Variable independiente x: Distancia al centro (km)
x = np.arange(1, 21).reshape(-1, 1)
print(x)

# Variable dependiente y: Precio (miles de $)
y = np.array([340, 320, 290, 260, 230,
              210, 220, 245, 275, 305,
              330, 350, 365, 375, 380,
              385, 390, 395, 398, 400])
print(y)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Definicion del pipeline de regresion lineal
pipeline = Pipeline(steps=[
    ("scaler", StandardScaler()),
    ("regressor", LinearRegression())
])

# Entrenamiento del modelo
pipeline.fit(x, y)

# Prediccion del modelo
y_pred = pipeline.predict(x)

plot_model(
    xlabel="Distancia al centro (Km)",
    ylabel="Precio (miles de $)",
    tittle="Estimaci贸n de precio de una casa (regresi贸n lineal simple)"
)
```
